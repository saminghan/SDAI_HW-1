```python
import numpy as np
from numpy import append, cos, linspace, pi, sin, zeros
import matplotlib.pyplot as plt
# Hipo parameters
parameters = [50 - 15j, 5 + 2j, -10 - 10j, -14 - 60j, -15 + 30j] 
#the fourier coefficients
npar = 6
p = parameters

Cx = zeros((npar,), dtype='complex') #shape = (npar, )
Cy = zeros((npar,), dtype='complex')

Cx[1] = p[0].real * 1j
Cy[1] = p[3].imag + p[0].imag * 1j

Cx[2] = p[1].real * 1j
Cy[2] = p[1].imag * 1j

Cx[3] = p[2].real
Cy[3] = p[2].imag * 1j

Cx[5] = p[3].real

t=linspace(0.6 + 1.3 * pi, 4 * pi + 0.9 * pi, 1000)

def fourier(t, C):
    f = zeros(t.shape)  #len(f) = len(t)
    for k in range(len(C)):
        f += C.real[k] * cos(k * t) + C.imag[k] * sin(k * t)
    return f
Cy1 = zeros((2,), dtype='complex')
Cy2 = zeros((3,), dtype='complex')
Cy3 = zeros((4,), dtype='complex')

Cy1[1] = p[3].imag + p[0].imag * 1j

Cy2[1] = p[3].imag + p[0].imag * 1j
Cy2[2] = p[1].real * 1j

Cy3[1] = p[3].imag + p[0].imag * 1j
Cy3[2] = p[1].real * 1j
Cy3[3] = p[2].imag * 1j

fx1 = fourier(t, Cy1)
fx2 = fourier(t, Cy2)
fx3 = fourier(t, Cy3)

fig, axs = plt.subplots(2,2, constrained_layout=True)
axs[0,0].plot(t, fx1, '-')
axs[0,0].set_title('The Hippo x-axis for k=1')
axs[0,0].set_xlabel('angle(rad)')
axs[0,0].set_ylabel('values')
fig.suptitle('The Curve of Hippo on X axes', fontsize=16)

axs[0,1].plot(t, fx2, 'r')
axs[0,1].set_title('The Hippo x-axis for k=2')
axs[0,1].set_xlabel('angle(rad)')
axs[0,1].set_ylabel('values')

axs[1,0].plot(t, fx3, 'b')
axs[1,0].set_title('The Hippo x-axis for k=3')
axs[1,0].set_xlabel('angle(rad)')
axs[1,0].set_ylabel('values')

axs[1,1].plot(t, fx1, '-', t,fx2,'r',t,fx3,'b')
axs[1,1].set_title('The Hippo x-axis for k=1,2,3')
axs[1,1].set_xlabel('angle(rad)')
axs[1,1].set_ylabel('values')
```

```python
Cx1 = zeros((2,), dtype='complex')
Cx2 = zeros((3,), dtype='complex')
Cx3 = zeros((4,), dtype='complex')
Cx5 = zeros((6,), dtype='complex')

Cx1[1] = p[0].real * 1j

Cx2[1] = p[0].real * 1j
Cx2[2] = p[1].real * 1j  

Cx3[1] = p[0].real * 1j
Cx3[2] = p[1].real * 1j    
Cx3[3] = p[2].real

Cx5[1] = p[0].real * 1j
Cx5[2] = p[1].real * 1j    
Cx5[3] = p[2].real
Cx5[5] = p[3].real

fy1 = fourier(t, Cx1)
fy2 = fourier(t, Cx2)
fy3 = fourier(t, Cx3)
fy5 = fourier(t, Cx5)

fig, axs = plt.subplots(2,2, constrained_layout=True)
axs[0,0].plot(t, fy1, '-')
axs[0,0].set_title('The Hippo y-axis for k=1')
axs[0,0].set_xlabel('angle(rad)')
axs[0,0].set_ylabel('values')
fig.suptitle('The Curve of Hippo on Y axes', fontsize=16)

axs[0,1].plot(t, fy2, 'r')
axs[0,1].set_title('The Hippo y-axis for k=2')
axs[0,1].set_xlabel('angle(rad)')
axs[0,1].set_ylabel('values')

axs[1,0].plot(t, fy3, 'b')
axs[1,0].set_title('The Hippo y-axis for k=3')
axs[1,0].set_xlabel('angle(rad)')
axs[1,0].set_ylabel('values')

axs[1,1].plot(t, fy5, 'g')
axs[1,1].set_title('The Hippo y-axis for k=5')
axs[1,1].set_xlabel('angle(rad)')
axs[1,1].set_ylabel('values')
```
